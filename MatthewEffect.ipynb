# 3. Prepare Data

```python
me = df.copy()
```

```python
# Cambiar columna a formato de fecha pandas
me['Birth Date'] = pd.to_datetime(me['Birth Date'])
```

```python
# Obtener los meses
me['Birth Date'].dt.to_period('M')
```

```python
me['Birth Date'].dt.strftime('%m/%Y')
```

```python
me['yyyy'] = pd.to_datetime(me['Birth Date']).dt.year
me['mm'] = pd.to_datetime(me['Birth Date']).dt.month
```

```python
me
```

```python
clean = me.groupby('mm')['No.'].nunique().to_frame()
clean = clean.reset_index()
```

```python
clean
```

# Analyze Data

```python
clean
```

```python
clean['mm'] = clean['mm'].replace([1,2,3,4,5,6,8,9,10,12],
                                  ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Agosto", "Septiembre", "Octubre", "Diciembre"])
```

```python
from plotly.graph_objs import *
fig = px.bar(clean, x='mm', y='No.', title="Matthew Effect", labels={'mm': 'Mes de Nacimiento', 'No.':'# de Jugadores'})

fig.show()
```
